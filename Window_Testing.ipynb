```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df = pd.read_csv("Catch22_Windows.csv")
df.drop(columns = "Unnamed: 0", inplace = True)
```

```python
df[df["Window_Size"] ==60]["DN_HistogramMode_5"]
```

```python
import matplotlib.pyplot as plt
import pandas as pd
window_sizes = [60, 300, 900, 1800, 3600]
masks = {size: df["Window_Size"] == size for size in window_sizes}
plt.figure(figsize=(10, 6)) 
for column in df.columns:
    if column != 'Window_Size':
        plt.figure()
        plt.title(f'Plot for {column}')
        for size in window_sizes:
            plt.plot(df[masks[size]][column].reset_index(drop=True), label=f'Window Size: {size}')
        plt.xlabel('Index')
        plt.ylabel(column)
        plt.legend()
        plt.show()
```
